```python
content="""
1. What is Simple Linear Regression?
•	A regression technique that models the relationship between one independent variable (X) and one dependent variable (Y) using a straight line:
Y=mX+cY = mX + cY=mX+c

2. What are the key assumptions of Simple Linear Regression?
1.	Linearity → Relationship between X and Y is linear.
2.	Independence → Observations are independent of each other.
3.	Homoscedasticity → Constant variance of residuals.
4.	Normality of residuals → Errors are normally distributed.

3. What does the coefficient m represent in Y = mX + c?
•	Slope (m) → Change in Y for a one-unit increase in X.
Example: If m = 2, then for each additional unit of X, Y increases by 2 (on average).

4. What does the intercept c represent in Y = mX + c?
•	Intercept (c) → Predicted value of Y when X = 0.

5. How do we calculate the slope m in Simple Linear Regression?

•	Numerator: covariance between X and Y.
•	Denominator: variance of X.

6. What is the purpose of the least squares method in Simple Linear Regression?
•	To minimize the sum of squared errors (differences between actual Y and predicted Y).
•	Ensures the best-fitting regression line.

7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?
•	R² = proportion of variance in Y explained by X.
•	Ranges 0 → 1:
o	R² = 0 → model explains nothing.
o	R² = 1 → perfect fit.
•	Example: R² = 0.8 → 80% of variance in Y is explained by X.
8. What is Multiple Linear Regression?
•	A regression technique that models the relationship between multiple independent variables (X₁, X₂, …, Xn) and a dependent variable (Y):
Y=b0+b1X1+b2X2+...+bnXn

9. Main difference between Simple and Multiple Linear Regression?
•	Simple → 1 independent variable.
•	Multiple → 2 or more independent variables.

10. Key assumptions of Multiple Linear Regression:
1.	Linearity between predictors and target.
2.	Independence of errors.
3.	Homoscedasticity (equal variance of errors).
4.	Normality of residuals.
5.	No multicollinearity (predictors shouldn’t be highly correlated).

11. What is heteroscedasticity? How does it affect results?
•	Heteroscedasticity → Non-constant variance of errors across values of X.
•	Effect:
o	Leads to biased standard errors.
o	Makes hypothesis tests (t, F tests) unreliable.

12. How can you improve a Multiple Linear Regression model with high multicollinearity?
•	Techniques:
o	Remove highly correlated predictors.
o	Use PCA (Principal Component Analysis).
o	Apply regularization (Ridge or Lasso regression).
o	Variance Inflation Factor (VIF) check.

13. Common techniques for transforming categorical variables for regression models:
1.	One-Hot Encoding → Creates dummy variables.
2.	Label Encoding (for ordinal data).
3.	Frequency/Target Encoding.


14. Role of interaction terms in Multiple Linear Regression:
•	Interaction terms capture situations where effect of one variable depends on another.
•	Example: Salary ~ Education + Experience + (Education × Experience).

15. Interpretation of intercept in Simple vs Multiple Linear Regression:
•	Simple → Value of Y when X = 0.
•	Multiple → Value of Y when all X’s = 0 (sometimes meaningless if 0 is not realistic).

16. What is the significance of the slope in regression analysis, and how does it affect predictions?
•	Slope = rate of change of Y with respect to X.
•	If slope = 3 → for every 1 unit increase in X, Y increases by 3 (keeping other factors constant).
•	It determines direction (positive/negative) and magnitude of effect.

17. How does the intercept in a regression model provide context for the relationship between variables?
•	Intercept = expected value of Y when all predictors = 0.
•	Context:
o	In Simple Regression → baseline value of Y at X = 0.
o	In Multiple Regression → baseline Y when all X’s = 0.
•	Sometimes has no practical meaning (e.g., height at age 0).

18. What are the limitations of using R² as a sole measure of model performance?
•	Always increases when more predictors are added (even useless ones).
•	Does not measure causation, only fit.
•	Does not detect overfitting.
•	Cannot compare models with different dependent variables.

19. How would you interpret a large standard error for a regression coefficient?
•	Means high variability/uncertainty in the estimate of that coefficient.
•	Suggests coefficient may not be statistically significant.
•	Could be due to:
o	Multicollinearity.
o	Small sample size.
o	Noise in data.


20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?
•	Identification:
o	Plot residuals vs predicted values.
o	If residuals fan out (cone shape), variance is not constant.
•	Importance:
o	Leads to biased standard errors → wrong p-values & confidence intervals.
o	Reduces reliability of hypothesis testing.

21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?
•	High R² → model explains variance well.
•	Low adjusted R² → many predictors don’t contribute meaningfully (overfitting).
•	Adjusted R² penalizes for irrelevant predictors.

22. Why is it important to scale variables in Multiple Linear Regression?
•	Prevents predictors with large scales (e.g., income in $ vs age in years) from dominating.
•	Helps in regularization methods (Ridge, Lasso).
•	Improves numerical stability of algorithms.

23. What is polynomial regression?
•	A form of regression where the relationship between X and Y is modeled as an nth-degree polynomial.
Y=b0+b1X+b2X2+...+bnXn

24. How does polynomial regression differ from linear regression?
•	Linear regression → straight line.
•	Polynomial regression → curve (can capture non-linear relationships).
•	Note: Still linear in coefficients, but non-linear in variables.

25. When is polynomial regression used?
•	When data shows a curved trend.
•	Example: growth curves, quadratic effects, learning curves.

26. What is the general equation for polynomial regression?
For degree n:
Y=b0+b1X+b2X2+...+bnXn

27. Can polynomial regression be applied to multiple variables?
•	Yes → Polynomial Multiple Regression.
Y=b0+b1X1+b2X12+b3X2+b4X22+...
•	Often used with interaction terms too.

28. What are the limitations of polynomial regression?
•	Prone to overfitting if degree is too high.
•	Poor extrapolation outside training range.
•	Computationally expensive for large datasets.
•	Multicollinearity increases with higher powers.

29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?
•	Cross-validation (k-fold).
•	Adjusted R².
•	AIC/BIC (Information criteria).
•	Validation set error (test MSE).

30. Why is visualization important in polynomial regression?
•	Helps check if polynomial curve fits the trend properly.
•	Detects overfitting (too wiggly curve).
•	Shows relationship between degree of polynomial and model complexity.
"""
print(content)
```

```python
#31. How is polynomial regression implemented in Python?
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1,2,3,4,5,6]).reshape(-1, 1)
y = np.array([2,5,7,10,15,20])

# Transform features into polynomial (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit model
model = LinearRegression()
model.fit(X_poly, y)

# Predictions
y_pred = model.predict(X_poly)

# Visualization
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.show()
```
